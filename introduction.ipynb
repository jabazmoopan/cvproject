```python
img_gray=cv2.imread('/Users/computer vision/open cv/shah-rukh-khan-helps-a-girl-adjust-her-handbag-melting-the-entire-internet-at-once-netizens-label-him-a-gentleman-watch-01.jpg',cv2.IMREAD_GRAYSCALE)
```

```python
cv2.imshow("grey_image",img_gray)
cv2.imshow('original_image',img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
(thresh,binary_img)=cv2.threshold(img_gray,128,255,cv2.THRESH_BINARY)
```

```python
edge_img=cv2.Canny(img,128,255)
cv2.imshow('edge_image',edge_img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt
plt.imshow(img)
```

```python
conv_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(conv_img)
```

```python
conv_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
rect=cv2.rectangle(conv_img,pt1=(450,10),pt2=(800,350),color=(255,0,0),thickness=5)
plt.imshow(rect)
```

```python
conv_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
cir=cv2.circle(conv_img,center=(650,200),radius=150,color=(0,255,0),thickness=10)
plt.imshow(cir)
```

```python
conv_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
text=cv2.putText(conv_img,text="srk",org=(0,150),fontFace=cv2.FONT_HERSHEY_COMPLEX,fontScale=6,color=(0,255,0),thickness=10)
plt.imshow(text)
```
